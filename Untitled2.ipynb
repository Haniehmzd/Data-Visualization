```python
import pandas as pd

# Load master dataset
df = pd.read_csv("Master_Dataset_Final.csv")

# Select one y
YEAR = 2014
df_year = df[df["Year"] == YEAR]


df_map = df_year[["Country Code", "CO2"]]

df_map = df_map.dropna(subset=["CO2"])

#TAb
output_file = "co2_map_2014.csv"
df_map.to_csv(output_file, index=False)

print(f"File saved: {output_file}")
```
